# *some definitions of AI agents:*

An llm with tools in a loop to achieve a goal


# Agentic system

**Anthropic distinguish two types**


**Workflows:** A system where LLMS and tools are orchestrated through predefined code paths.

**Agents:** Are dynamic systems where LLMS direct their own processes and tool usage, maintaing control over how they accomplish tasks. 

# Anthropic designed 5 workflow patterns

**1 - Prompt chaining**

**2 - Routing**

**3 - Parallelization**

**4 - Orchestrator-worker**

**5 - Evaluator-Optimizar**

# 1 - Prompt chaining

**Decompose into fixed sub-tasks**

**How it works:** Series of LLms calls where each output will be taken as an input to the another call



# 2 - Routing

**Direct an input into a specialized sub-task, ensuring sepration of concerns**

**How it works:** An input will be given to an LLM router which means he will take the input and make a decision to route it to which LLM (there will be multiple LLM models)

# 3 - Parallelization

**Breaking down tasks ans running multiple subtasks concurrently.**

**How it works:** Write a python code to break down a huge task into small tasks and give them to LLMs to run them concurrently and their output is then passed to python code to aggregate the result and produce the final output

# 4 - Orchestrator-worker

**Complex tasks are broken down dynamically and combined.**

**How it works:** A **LLM decide** how to break down a huge task into small tasks and give them to LLMs to run them concurrently and their output is then passed to another **LLM to aggregate** the result and produce the final output

# 5 - Evaluator-Optimizar

**LLM output is evaluated by another in loop untill it is accepted**

**How it works:** A LLM generate an output, this output is evaluted by antother LLM and if it is rejected it will return it back to the first LLM with some notes and that loop will continuously running untill the evaluator LLM accept it then the output will passed to the user.

# Agents

**1 - Open-ended**
**2 - Feedback loops**
**3 - No fixed path**

**How it works:** An LLM uses tools within a dynamic feedback loop to achieve a specific goal. All execution decisions are taken autonomously by the LLM (no fixed path) until it determines that the final output is correct, at which point the result is passed back to the user.
